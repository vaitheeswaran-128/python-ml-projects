# BIVARIATE

```python
grouped = df.groupby("cancellation_policy").size().reset_index(name="counts")

plt.figure(figsize=(8, 8))

explode = [0.1] * len(grouped)
plt.pie(
    grouped["counts"],                     
    labels=grouped["cancellation_policy"],   
    autopct='%1.1f%%',                       
    startangle=140,                         
    explode=explode,                        
    colors=sns.color_palette("pastel")       
)

plt.title("Count of Cancellation Policy Types", fontsize=16)
plt.savefig("piechart.png",format="png")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

room_accommodates = df.groupby("room_type")["accommodates"].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=room_accommodates, x="room_type", y="accommodates", palette="pastel")

ax = plt.gca()

for bar in ax.patches:
    x = bar.get_x() + bar.get_width() / 2  
    y = bar.get_height()  
    ax.text(x, y, f'{y:.0f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45)
plt.title("Total Accommodates per Room Type")
plt.xlabel("Room Type")
plt.ylabel("Total Accommodates")
plt.savefig("barchart_2.png",format="png")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

for room_type, group in df.groupby("room_type"):
    sns.kdeplot(data=group, x="accommodates", fill=True, label=room_type)

plt.title("KDE Plot of Accommodates by Room Type", fontsize=14)
plt.xlabel("Number of People Accommodated", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(title="Room Type")

plt.savefig("KDEplot.png",format="png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x="room_type", y="review_scores_rating", palette="muted", inner="quartile")
plt.title("Room Type-wise Review Ratings (Violin Plot)")
plt.xlabel("Room Type")
plt.ylabel("Review Scores Rating")
plt.xticks(rotation=45)
plt.savefig("violinplot.png",format="png")
plt.show()
```
